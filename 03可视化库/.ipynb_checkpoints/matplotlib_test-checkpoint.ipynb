```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import numpy as np
import os
import pandas as pd
import requests
plt.rc('font', family='SimHei', size=13)

PATH = 'F:/Git/ML_Python/02iris/'
os.chdir(PATH)
df = pd.read_csv(PATH + 'iris.data',names=['花萼长度','花萼宽度','花瓣长度','花瓣宽度','类别'])
df.head()
```

```python
fig, ax = plt.subplots(2,2,figsize=(6,4))
ax[0][0].hist(df['花萼长度'], color ='green')
ax[0][0].set_ylabel('count',fontsize=12)
ax[0][0].set_xlabel('length',fontsize=12)
ax[0][0].set_title('Iris Petal Lenth', fontsize=14, y=1.01)

ax[0][1].hist(df['花萼宽度'], color ='green')
ax[0][1].set_ylabel('count',fontsize=12)
ax[0][1].set_xlabel('width',fontsize=12)
ax[0][1].set_title('Iris Petal Width', fontsize=14, y=1.01)

ax[1][0].hist(df['花瓣长度'], color ='green')
ax[1][0].set_ylabel('count',fontsize=12)
ax[1][0].set_xlabel('length',fontsize=12)
ax[1][0].set_title('Iris Sepal Length', fontsize=14, y=1.01)

ax[1][1].hist(df['花瓣宽度'], color ='green')
ax[1][1].set_ylabel('count',fontsize=12)
ax[1][1].set_xlabel('width',fontsize=12)
ax[1][1].set_title('Iris Sepal Width', fontsize=14, y=1.01)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(df['花瓣宽度'],df['花瓣长度'],color='red')
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatterplot')
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(df['花瓣长度'],color='blue')
ax.set_xlabel('num')
ax.set_ylabel('Petal Length')
ax.set_title('Petal length plot')
```

```python
fig, ax = plt.subplots(figsize = (6,6))
bar_width = 0.8
labels = [x for x in df.columns if '长度' in x or '宽度' in  x]
ver_y = [df[df['类别']=='Iris-versicolor'][x].mean() for x in  labels]
vir_y = [df[df['类别']=='Iris-virginica'][x].mean() for x in  labels]
set_y = [df[df['类别']=='Iris-setosa'][x].mean() for x in  labels]
x = np.arange(len(labels))
ax.bar(x, vir_y, bar_width, bottom = set_y, color='darkgrey')
ax.bar(x, set_y, bar_width, bottom = ver_y, color='white')
ax.bar(x, ver_y, bar_width, color='black')
ax.set_xticks(x + (bar_width/20))
ax.set_xticklabels(labels, rotation=-50, fontsize=12)
ax.set_title('Mean Feature Measurement By Class', y=1.01)
ax.legend(['Virginica','Setosa','Versicolor'])
```
